**first code**

```python
import numpy as np
```

```python
#given data
x=np.array([1.0,2.0]) #ip
y=1  #target
w=np.array([0.2,-0.5]) #intial weights
b=0.1 #bias
eta=0.1 #learning rate
```

```python
#run for 3 epochs
for epoch in range(1,4):
  print(f"\nEpoch {epoch}:") #f is for teeling it is string format
  net_ip=np.dot(w,x)+b
  y_pred=1 if net_ip>0 else 0
  error=y-y_pred

  print(f" net input={net_ip:.2f}")
  print(f" pred_op={y_pred}")
  print(f"error={error}")


  #update rule
  w=w+eta*error*x
  b=b+eta*error


  print(f"updated weights={w}")
  print(f"updated bias={b:.2f}")

```

```python
# testing phase
x_test=np.array([2.0,1.0])
net_test=np.dot(w,x_test)+b
y_test_pred=1 if net_test>0 else 0

print("\n testing phase")
print(f"test inputs={x_test}")
print(f" net_input={net_test:.2f}")
print(f"prediction output={y_test_pred}")
```
